```python
class vectoresR7:
    def __init__(self,L):
        self.L=L
    def __add__(self,v1):
        L=[self.L[i]+v1.L[i] for i in range (7)]
        return vectoresR7(L)
    def __repr__(self):
        return repr(self.L)
```

```python
class cajero:
    def __init__(self,billete_10,billete_20):
        self.billete_10=billete_10
        self.billete_20=billete_20   
    def retiro(self,cantidad):
        cant=cantidad/10000
        if cantidad%10000!=0:
            return 'ingrese un multiplo de 10000'
        nc2= cant//2
        pd2= min(nc2,self.billete_20)
        
        restante= cant-pd2*2
       
        nc1= restante
        pd1= min(nc1,self.billete_10)
        restante-= pd1
        if restante==0:
            self.billete_10-=pd1
            self.billete_20-=pd2
            return f'le entrego {pd2} billetes de 20.000 y {pd1} billetes de 10.000'
        else:
            return f'el cajero no tiene esa cantidad de dinero, lo siento..'
        
```

```python
p1=cajero(10,10)
```

```python
print(p1.retiro(20000))
print(p1.billete_10,p1.billete_20)
```
