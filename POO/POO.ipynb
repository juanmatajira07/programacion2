```python
#sumando vectores
class vectoresR2:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def sumarvectores(self,otro_vector):
        x1=self.x + otro_vector.x
        y1=self.y + otro_vector.y
        return vectoresR2(x1,y1)
    
    #otra manera haciendolo solo con suma
    def __add__(self,otro_vector):
        x1=self.x + otro_vector.x
        y1=self.y + otro_vector.y
        return vectoresR2(x1,y1)
    
    def restarvectores(self,otro_vector):
        x1=self.x - otro_vector.x
        y1=self.y - otro_vector.y
        return vectoresR2(x1,y1)
    
    def multiplicacion_escalar(self,escalar):
        x1=self.x * escalar
        y1=self.y * escalar
        return vectoresR2(x1,y1)
    
    def producto_punto(self,otro_vector):
        x1=self.x * otro_vector.x
        y1=self.y * otro_vector.y
        z1=x1+y1
        return z1

    def __repr__(self):
        return f'({self.x},{self.y})'
V1=vectoresR2(2,4)
V2=vectoresR2(3,3)
```
