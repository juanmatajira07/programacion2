```python
class persona:
    def __init__(self,promedio,nombre):
        self.promedio=promedio
        self.nombre=nombre
    def saludo(self):
        return f'mi nombre es {self.nombre}'
    def aprobar(self,nota_minima):
        return self.promedio>=nota_minima
```

```python
p1=persona(3.5,'juan')
p1.saludo(),p1.aprobar(3.5)
```
