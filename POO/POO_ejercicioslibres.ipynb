```python
class persona:
    def __init__(self,promedio,nombre):
        self.promedio=promedio
        self.nombre=nombre
    def saludo(self):
        return f'mi nombre es {self.nombre}'
    def aprobar(self,nota_minima):
        return self.promedio>=nota_minima
```

```python
p1=persona(3.5,'juan')
p1.saludo(),p1.aprobar(3.5)
```

```python
class cuenta:
    def __init__(self,titular,saldo):
        self.titular=titular
        self.saldo=saldo
    def depositar(self,cantidad):
        self.saldo+=cantidad
        return f'ingreso {cantidad} pesos, su saldo es de{self.saldo}'
    def retirar(self,cantidad):
        self.saldo-=cantidad
        return f'retiro {cantidad} pesos, su saldo es de {self.saldo}'
    def consultar_saldo(self):
        return f'su saldo es de {self.saldo}'
p2=cuenta('juan',500)
    
```

```python
p2.retirar(200)
```

```python
class cajero:
    def __init__(self,n1,n2,n3):
        self.n1=n1
        self.n2=n2
        self.n3=n3
    def retiro(self,cantidad):
        cant=cantidad/10000
        if cantidad%10000!=0:
            return 'ingrese un multiplo de 10000'
        nc3=cant//5
        pd3=min(nc3,self.n3)
       
        restante=cant-pd3*5
       
        nc2= restante//2
        pd2= min(nc2,self.n2)
        
        restante-=pd2*2
       
        nc1= restante
        pd1= min(nc1,self.n2)
        restante-= pd1
        if restante==0:
            self.n1-=pd1
            self.n2-=pd2
            self.n3-=pd3
            return f'le entrego {pd3} billetes de 50.000, {pd2} billetes de 20.000 y {pd1} billetes de 10.000'
        else:
            return f'el cajero no tiene esa cantidad de dinero, lo siento..'
    def consignacion(self,n1,n2,n3):
        self.n1+=n1
        self.n2+=n2
        self.n3+=n3
        return f'ahora tiene {self.n1} billetes de 10, {self.n2} billetes de 20 y {self.n3} billetes de 50'
p1=cajero(10,10,10)
```

```python
p1.retiro(70000)
```

```python
class recta:
    def __init__(self,x1,y1,x2,y2):
        self.x1=x1
        self.y1=y1
        self.x2=x2
        self.y2=y2
    def pendiente(self):
        return self.y2-self.y1/self.x2-self.x1
```
